```python

def spruce(size):
    print("a spruce!")  # Başlık yazdır

    for i in range(1, size * 2, 2):  # 1, 3, 5, ..., (size*2 - 1) olacak şekilde artar
        print(" " * ((size * 2 - 1 - i) // 2) + "*" * i)  # Boşluklar + Yıldızlar

    print(" " * (size - 1) + "*")  # Gövde yazdır

# Örnek kullanım:
spruce(5)
```

```python
def ilkKelime(cumle):
    return cumle.split()[0]

def ikinciKelime(cumle):
    return cumle.split()[1]

def sonKelime(cumle):
    return cumle.split()[-1]

# Örnek kullanım:
cumle = "örnek bir python cümlesi"
print(ilkKelime(cumle))
print(ikinciKelime(cumle))
print(sonKelime(cumle))
```
